```python
import videodataset
import numpy as np
import matplotlib.pyplot as plt
import torch

from torch.utils.data import DataLoader
```

```python
video_list_paths = "videolist/UCF-101/train_data_list.txt"
# get video list from video_list_paths
with open(video_list_paths, 'r') as f:
    video_list = [line.strip() for line in f.readlines()]
```

```python
train_dataset = videodataset.VideoDataset(video_list,
                                          seq_length=11,
                                          img_size=128,
                                          img_channel=3,
                                          mode="train",
                                          mask_probability=0)
train_loader = DataLoader(dataset=train_dataset, batch_size=4, shuffle=True, num_workers=1)
```

```python
def pixelDownShuffle(img_tensor, patch_size):
    shape = np.shape(img_tensor)
    print(shape)
    batch_size = shape[0]
    seq_length = shape[1]
    img_channel = shape[2]
    img_size = shape[3]
    print(batch_size)
    
    img_tensor = np.transpose(img_tensor, [0,1,3,4,2])
    a = np.reshape(img_tensor, [batch_size, seq_length,
                                   int(img_size/patch_size), patch_size,
                                   int(img_size/patch_size), patch_size, img_channel])
    b = np.transpose(a, [0,1,2,4,3,5,6])
    patch_tensor = np.reshape(b, [batch_size, seq_length,
                                  int(img_size/patch_size),
                                  int(img_size/patch_size), patch_size*patch_size*img_channel])
    
    patch_tensor = np.transpose(patch_tensor, [0,1,4,2,3])
    
    return patch_tensor
```

```python
plt.figure(figsize=(15, 10))
for vid_path, seq, seq_gt in train_loader:
    print(vid_path)
    patch_tensor = pixelDownShuffle(seq, 4)
    print(np.shape(patch_tensor))
#     print('Size of seq:', seq.size())
    for index, img in enumerate(patch_tensor.numpy()[0]):
        plt.subplot(4, 6, index+1)
        plt.imshow(img[0], cmap='gray')
#         plt.imshow(np.transpose(img, (1, 2, 0)))
    seq_gt = torch.flip(seq_gt, (1, ))
    for index, img in enumerate(seq_gt.numpy()[0]):
        plt.subplot(4, 6, index+1+12)
        plt.imshow(np.transpose(img, (1, 2, 0)))
    break
```
