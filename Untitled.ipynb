```python
import videodataset
from model.model_factory import Model
from utils.metrics import init_meters

import numpy as np
import matplotlib.pyplot as plt
import argparse
import math
import random
import os
import cv2
from tqdm import tqdm

import torch
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
```

### I / O parameter

```python
def process_command():
    parser = argparse.ArgumentParser()
    
    parser.add_argument('--dataset_name', default='vimeo90K')
    parser.add_argument('--valid_data_paths', default='videolist/Vimeo90K/test_list.txt')
    parser.add_argument('--gen_frm_dir', default='../MVFI_output/results/test')
    parser.add_argument('--model_name', default='BiLSTM')
    parser.add_argument('--seq_length', default=7, type=int)
    parser.add_argument('--img_size', default=256, type=int)
    parser.add_argument('--img_channel', default=3, type=int)
    parser.add_argument('--num_hidden', default='64,64,64,64')
    parser.add_argument('--filter_size', default=5, type=int)
    parser.add_argument('--stride', default=1, type=int)
    parser.add_argument('--patch_size', default=4, type=int)
    parser.add_argument('--batch_size', default=32, type=int)
    parser.add_argument('--LSTM_pretrained', default='')
    parser.add_argument('--lr', default=0.001, type=float)
    parser.add_argument('--loss', default='L1+L2')
    
    return parser.parse_args(args=[
        '--dataset_name', 'base_dataset',
        '--valid_data_paths', 'videolist/UCF-101/train_data_list.txt',
        '--gen_frm_dir', '../MVFI_output/results/test',
        '--model_name', 'BiLSTM',
        '--seq_length', '7',
        '--img_size', '224',
        '--img_channel', '3',
        '--num_hidden', '32,32,32,32',
        '--filter_size', '5',
        '--stride', '1',
        '--patch_size', '4',
        '--batch_size', '32',
        '--LSTM_pretrained', '',
        '--lr', '0.001',
        '--loss', 'L1+L2'
    ])
    
args = process_command()
```

### Loading LSTM model

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print("Loading LSTM model...", end='')   
LSTM = Model(args, device)
LSTM.network.eval()
print("...OK")
```

```python
# get video list from video_list_paths
with open(args.valid_data_paths, 'r') as f:
    valid_video_list = [line.strip() for line in f.readlines()]
```

```python
checkpoint_path = '../checkpoint_199.tar'
checkpoint = torch.load(checkpoint_path)
model_state_dict = checkpoint['model_state_dict']
optimizer_state_dict = checkpoint['optimizer_state_dict']
pretrained_epoch = checkpoint['epoch']
mask_probability = checkpoint['mask_probability']
# model loading weight
LSTM.load_checkpoint(model_state_dict, optimizer_state_dict)
```

```python
valid_dataset = videodataset.VideoDataset(valid_video_list, args.dataset_name,
                                          seq_length=args.seq_length, 
                                          img_size=args.img_size, 
                                          img_channel=args.img_channel, 
                                          mode="valid", 
                                          mask_probability=0)
valid_loader = DataLoader(dataset=valid_dataset, batch_size=args.batch_size, shuffle=False, num_workers=12)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10, 10))
for vid_path, seq, seq_gt in valid_loader:
    print(np.shape(seq))
    imgs = seq[0]
    for i, img in enumerate(imgs):
        plt.subplot(1, 7, i+1)
        plt.imshow(img.permute(1, 2, 0))
    break
```

```python
print("Testing...")
psnrs, ssims = init_meters()
for vid_path, seq, seq_gt in tqdm(valid_loader):
    LSTM.test(vid_path, args.gen_frm_dir, seq, seq_gt, 0, psnrs, ssims)
```
